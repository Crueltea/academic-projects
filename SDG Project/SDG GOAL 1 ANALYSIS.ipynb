### Creating New Dataframes for tabs 1-12

For this section, we separate the dataframes by a creating a new one. One with regions, and the other with provinces.

```python
dfset1 = [
    tab1_cleaned_df, tab2_cleaned_df, tab3_cleaned_df, tab4_cleaned_df, 
    tab5_cleaned_df, tab6_cleaned_df, tab7_cleaned_df, tab8_cleaned_df, 
    tab10_cleaned_df, tab11_cleaned_df, tab12_cleaned_df
]

regions_to_extract = [
    "National Capital Region (NCR)", "Cordillera Administrative Region (CAR)", 
    "Region I (Ilocos Region)", "Region II (Cagayan Valley)", "Region III (Central Luzon)", 
    "Region IV-A (CALABARZON)", "MIMAROPA Region", "Region V (Bicol Region)", 
    "Region VI (Western Visayas)", "Negros Island Region (NIR)", "Region VII (Central Visayas)", 
    "Region VIII (Eastern Visayas)", "Region IX (Zamboanga Peninsula)", 
    "Region X (Northern Mindanao)", "Region XI (Davao Region)", "Region XII (SOCCSKSARGEN)", 
    "Region XIII (Caraga)", "Bangsamoro Autonomous Region in Muslim Mindanao (BARMM)"
]
```

```python
for i, df in enumerate(dfset1, start=1):
    df['Region/Province'] = df['Region/Province'].str.strip()
    selected_df = df[df['Region/Province'].isin(regions_to_extract)].copy()
    cleaned_df = df[~df['Region/Province'].isin(regions_to_extract)].copy()
    selected_df.rename(columns={'Region/Province': 'Region'}, inplace=True)
    cleaned_df.rename(columns={'Region/Province': 'Region'}, inplace=True)
    
    globals()[f'tab{i}_selected_regions_df'] = selected_df
    globals()[f'tab{i}_cleaned_df'] = cleaned_df
    print(f"Processed tab{i}_cleaned_df and tab{i}_selected_regions_df")
```

### Poverty Incidence and Subsistence

```python
def plot_distribution(df, column, title, ax): 
    sns.histplot(df[column], kde=True, ax=ax, color='red')
    ax.set_title(title)
    ax.set_xlabel("Incidence (%)")
    ax.set_ylabel("Frequency")

# Plot Poverty Incidence (Tabs 1 and 2)
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
fig.suptitle("Poverty Incidence Distributions (2018, 2021, 2023)", fontsize=16)

plot_distribution(tab1_selected_regions_df, 'Estimates 2018 (%)', "Poverty Incidence among Families (2018)", axes[0, 0])
plot_distribution(tab1_selected_regions_df, 'Estimates 2021 (%)', "Poverty Incidence among Families (2021)", axes[0, 1])
plot_distribution(tab1_selected_regions_df, 'Estimates 2023 (&)', "Poverty Incidence among Families (2023)", axes[0, 2])

plot_distribution(tab2_selected_regions_df, 'Estimates 2018 (%)', "Poverty Incidence among Population (2018)", axes[1, 0])
plot_distribution(tab2_selected_regions_df, 'Estimates 2021 (%)', "Poverty Incidence among Population (2021)", axes[1, 1])
plot_distribution(tab2_selected_regions_df, 'Estimates 2023 (&)', "Poverty Incidence among Population (2023)", axes[1, 2])

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

```python
# Plot Subsistence Incidence (Tabs 3 and 4)
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
fig.suptitle("Subsistence Incidence Distributions (2018, 2021, 2023)", fontsize=16)

plot_distribution(tab3_selected_regions_df, 'Estimates 2018 (%)', "Subsistence Incidence among Families (2018)", axes[0, 0])
plot_distribution(tab3_selected_regions_df, 'Estimates 2021 (%)', "Subsistence Incidence among Families (2021)", axes[0, 1])
plot_distribution(tab3_selected_regions_df, 'Estimates 2023 (&)', "Subsistence Incidence among Families (2023)", axes[0, 2])

plot_distribution(tab4_selected_regions_df, 'Estimates 2018 (%)', "Subsistence Incidence among Population (2018)", axes[1, 0])
plot_distribution(tab4_selected_regions_df, 'Estimates 2021 (%)', "Subsistence Incidence among Population (2021)", axes[1, 1])
plot_distribution(tab4_selected_regions_df, 'Estimates 2023 (&)', "Subsistence Incidence among Population (2023)", axes[1, 2])

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

```python
import matplotlib.cm as cm


# Calculate the average poverty incidence and sort by it
tab2_selected_regions_df['Average Incidence (%)'] = tab2_selected_regions_df[['Estimates 2018 (%)', 'Estimates 2021 (%)', 'Estimates 2023 (&)']].mean(axis=1)
sorted_df = tab2_selected_regions_df.sort_values(by='Average Incidence (%)', ascending=True)

# Normalize the average incidence values to a 0-1 range for the colormap
norm = plt.Normalize(sorted_df['Average Incidence (%)'].min(), sorted_df['Average Incidence (%)'].max())
colors = cm.RdYlGn_r(norm(sorted_df['Average Incidence (%)']))

# Plot the horizontal bar chart with gradient color
plt.figure(figsize=(10, 8))
plt.barh(sorted_df['Region'], sorted_df['Average Incidence (%)'], color=colors)
plt.xlabel("Average Poverty Incidence among Population (%)")
plt.title("Average Poverty Incidence among Population by Region (2018-2023)")
plt.colorbar(cm.ScalarMappable(norm=norm, cmap="Reds"), label="Incidence (%)")
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.tick_params(axis='y', which='both', length=0)
plt.show()
```

```python
top_10_most_poverty_2023 = tab2_cleaned_df.sort_values(by='Estimates 2023 (&)', ascending=False).head(10)
top_10_least_poverty_2023 = tab2_cleaned_df.sort_values(by='Estimates 2023 (&)', ascending=True).head(10)
```

```python
fixed_norm = plt.Normalize(0, 50)

colors_most = cm.RdYlGn_r(fixed_norm(top_10_most_poverty_2023['Estimates 2023 (&)']))

plt.figure(figsize=(10, 6))
plt.barh(top_10_most_poverty_2023['Region'], top_10_most_poverty_2023['Estimates 2023 (&)'], color=colors_most)
plt.xlabel("Poverty Incidence among Population (%)")
plt.title("Top 10 Provinces with the Most Poverty Incidence in 2023")
plt.gca().invert_yaxis()  # Invert y-axis for descending order
plt.colorbar(cm.ScalarMappable(norm=fixed_norm, cmap="RdYlGn_r"), label="Poverty Incidence (%)")
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.tick_params(axis='y', which='both', length=0)
plt.show()
```

```python
colors_least = cm.RdYlGn_r(fixed_norm(top_10_least_poverty_2023['Estimates 2023 (&)']))

plt.figure(figsize=(10, 6))
plt.barh(top_10_least_poverty_2023['Region'], top_10_least_poverty_2023['Estimates 2023 (&)'], color=colors_least)
plt.xlabel("Poverty Incidence among Population (%)")
plt.title("Top 10 Provinces with the Least Poverty Incidence in 2023")
plt.gca().invert_yaxis()  # Invert y-axis for ascending order
plt.colorbar(cm.ScalarMappable(norm=fixed_norm, cmap="RdYlGn_r"), label="Poverty Incidence (%)")

for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.tick_params(axis='y', which='both', length=0)
plt.show()
```

### Employment and Unemployment

```python
top_30_regions = tab17_cleaned_df.sort_values(by='Employment Rate 2022 (%)', ascending=False).head(15)

# Plot the horizontal bar chart
plt.figure(figsize=(12, 10))
bars = plt.barh(top_30_regions['Region/Province'], top_30_regions['Employment Rate 2022 (%)'], color='Green')
plt.xlabel("Employment Rate (%)")
plt.title("Top 15 Provinces/Cities with the Highest Employment Rate in 2022")
plt.gca().invert_yaxis()  # Invert y-axis to show highest rate at the top

# Add annotations for each bar
for bar in bars:
    plt.text(
        bar.get_width() + 0.5,  # Position to the right of the bar
        bar.get_y() + bar.get_height() / 2,  # Vertically centered on the bar
        f"{bar.get_width():.1f}%",  # Format the label as a percentage
        va='center', color='black'
    )

plt.tight_layout()  # Adjust layout to fit all labels
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```
